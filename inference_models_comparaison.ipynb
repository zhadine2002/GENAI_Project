# Test de géneration d'image avant Fine-tuning du modèle :stabilityai/stable-diffusion-xl-base-1.0

```python
from diffusers import DiffusionPipeline
import torch
```

```python
# Chargement du pipeline Stable Diffusion
model_id = "stabilityai/stable-diffusion-xl-base-1.0"
pipe = DiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe.to("cuda")
pipe.enable_attention_slicing()  # Réduction mémoire

tokenizer = pipe.tokenizer


```

```python
# Test de génération d'image
prompt = "A man standing wearing a blue suit smiling"
with torch.no_grad():
    result = pipe(prompt)
    image = result.images[0].convert("RGB")

display(image)
```
